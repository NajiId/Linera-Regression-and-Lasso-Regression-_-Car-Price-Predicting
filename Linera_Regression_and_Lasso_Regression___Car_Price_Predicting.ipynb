```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from  sklearn.linear_model import LinearRegression, Lasso
from sklearn import metrics
```

```python
data = pd.read_csv('car data.csv')
data.head()
```

```python
print(data['Fuel_Type'].value_counts(),'\n')
print(data['Seller_Type'].value_counts(),'\n')
print(data['Transmission'].value_counts())
```

Encoding Categorical Data

```python
data.replace({'Fuel_Type':{'Petrol':0, 'Diesel': 1, 'CNG':2}}, inplace=True)
data.replace({'Seller_Type':{'Dealer':0, 'Individual': 1}}, inplace=True)
data.replace({'Transmission':{'Manual': 0, 'Automatic': 1}}, inplace=True)
```

```python
X = data.drop(['Car_Name', 'Selling_Price'], axis=1, errors='ignore')
Y = data['Selling_Price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
```

```python
lin_rg_model = LinearRegression()
lasso_model = Lasso()
```

```python
X_train
```

```python
lin_rg_model.fit(X_train, Y_train)
```

```python
lin_rg_prediction = lin_rg_model.predict(X_test)
lin_reg_r2_score = metrics.r2_score(Y_test, lin_rg_prediction)
print(lin_reg_r2_score)
```

```python
lasso_model.fit(X_train, Y_train)
```

```python
lasso_model_prediction = lasso_model.predict(X_test)
lasso_r2_score = metrics.r2_score(Y_test, lasso_model_prediction)
print(lasso_r2_score)
```
